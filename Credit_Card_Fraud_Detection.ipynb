```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#load the dataset
df=pd.read_csv("creditcard.csv")
df.head()
```

```python
#summary of the DataFrame
df.info()
```

```python
df.shape
```

```python
#statistical summaries of numerical columns
df.describe()
```

```python
#Checking for missing values in each column
df.isna().sum()
```

```python
#Cleaning the dataset
# Dropping the rows with null values
df_cleaned = df.dropna()
df_cleaned.isna().sum()
```

```python
#pairwise correlation of all columns
corr=df_cleaned.corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='Blues',annot_kws={'size':4},fmt='.1f',cbar=True,square=True,)
plt.title("Correlation Matrix")
plt.show()
```

```python
# Counting the number of occurrences of each class (fraudulent or legitimate) in the Class column.
df_cleaned["Class"].value_counts()
#here 0->Legit
#here 1->Fraud
```

```python
legit=df_cleaned[df_cleaned.Class==0]
fraud=df_cleaned[df_cleaned.Class==1]
```

```python
legit.shape
```

```python
fraud.shape
```

```python
#Undersampling legit ones to fraud
legit_sample=legit.sample(n=492)
```

```python
#creating a new balanced dataset.
new_df=pd.concat([legit_sample,fraud],axis=0)
new_df
```

```python
new_df["Class"].value_counts()
```

```python
x=new_df.drop("Class",axis=1)
y=new_df.Class
print(x)
print(y)
```

```python
#split the data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#creating the model and training it
model=LogisticRegression()
model.fit(x_train,y_train)
print("The model is trained")
```

```python
#model evaluation
#Accuracy on training data
print("Accuracy on trained data: ",accuracy_score(y_train,model.predict(x_train)))

#Accuracy on testing data
print("Accuracy on test data: ",accuracy_score(y_test,model.predict(x_test)))
```
